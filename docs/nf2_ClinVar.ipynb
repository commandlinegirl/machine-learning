```python
import pandas as pd

df = pd.read_csv('/home/data/clinvar_result_combined_last.txt', delimiter="\t")

select_cols = df[['Name',
                  'Review status',
                  'AlleleID(s)',
                  'VariationID',
                  'Gene(s)',
                  'GRCh38Chromosome',
                  'GRCh38Location',
                  'AlleleID(s)',
                  'variant_type'
                 ]]

# strip "(Last reviewed.*" from each 'Clinical significance' record
significance = df['Clinical significance (Last reviewed)']
significance_stripped = significance.str.lower().str.replace(r'\(last reviewed.*', '')
select_cols = select_cols.assign(Significance = significance_stripped)

# group by
agg_funcs = {
             'Name': {'Names': [lambda i: ", ".join(i)]},
#              'variant_type': [lambda i: ", ".join(set(i))],
             'VariationID': {'Count': [lambda i: str(len([str(j) for j in i]))]}
            }

{'age':{'foo':'sum'},
     'height':{'mean':'mean', 'std':'std'}}

x = select_cols.groupby(['Significance', 'variant_type']).agg(agg_funcs)
x.columns = x.columns.droplevel(1)
x
```
